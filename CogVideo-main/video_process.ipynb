```python
import os
import sys
import glob

train_video_dirs = glob.glob(os.path.join("data", "typhoon_videos", "typhoon*"))

train_video_dirs
```

```python
def load_mp4_files(path):
    mp4_files = []
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith(".mp4"):
                mp4_files.append(os.path.join(root, file))
    return mp4_files

import torch
from torchvision.io import read_video, write_video

prompt = "typhoon"

for train_video_dir in train_video_dirs:
    print(f"Processing directory: {train_video_dir}")
    video_files = load_mp4_files(train_video_dir)
    video_files = sorted(video_files, key=lambda x: int(x.split("_")[-1].split(".")[0]))
    print(video_files)
    if len(video_files) == 0:
        print(f"No .mp4 files found in {train_video_dir}.")
        continue

    videos = []
    for video_file in video_files:
        try:
            video, audio, info = read_video(video_file)
            videos.append(video)
            print(f"Video shape: {video.shape}, Audio shape: {audio.shape}, Info: {info}")
        except Exception as e:
            print(f"Error reading {video_file}: {e}")
    
    combined_videos = []
    for video_p, video_n in zip(videos[:-1], videos[1:]):
        video_c = torch.cat((video_p, video_n), dim=0)[:49]
        combined_videos.append(video_c)
    
    save_paths = []

    for video_file_p, video_file_n in zip(video_files[:-1], video_files[1:]):
        video_file_split = video_file_p.split(os.sep)
        video_file_split[-2] = video_file_split[-2] + "_c"
        save_path = os.sep.join(video_file_split)
        save_paths.append(save_path)
    
    # save videos to save_paths
    for save_path, video in zip(save_paths, combined_videos):
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        try:
            write_video(save_path, video, fps=5)
            print(f"Saved combined video to {save_path}")
        except Exception as e:
            print(f"Error saving {save_path}: {e}")
```
